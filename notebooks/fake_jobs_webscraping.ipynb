```python
#extract_job extracts the job description from the 'Apply' page of the website
def extract_job_desc(page_url):
    response3=requests.get(page_url)
    soup3=BeautifulSoup(response3.text)
    job_desc=soup3.find(class_='content').findAll('p')[0].text.strip()
    return job_desc
```

```python

all_job_descriptions=[]
for j in range (len(urls)):
    temp=urls[j]
    all_job_descriptions.append(extract_job_desc(temp))
all_job_descriptions


#ajd=urls.apply(extract_job_desc)

```
